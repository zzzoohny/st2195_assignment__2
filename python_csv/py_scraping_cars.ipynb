```python
# Import/Load Libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
import os
#import urllib.request
```

```python
# Load in the html


url = 'https://en.wikipedia.org/wiki/Comma-separated_values'
r = requests.get(url)
print(type(r))

html_content = r.text

#.text needed to change them into text format
soup = BeautifulSoup(html_content, 'html.parser')
print(help(BeautifulSoup))
```
